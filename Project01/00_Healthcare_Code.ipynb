```python
import pandas as pd
import numpy as np 

# Settings seed for reproductibility 
np.random.seed(42)

#Generating  1,000 rows of data
data= {
    'Patient_ID': range(101,1101),
       'Name': [f"Patient_{i}" for i in range(1000)],
       'Age':np.random.randint(-5,110, size =1000), #Includes messy data (negative age) 
       'Gender': np.random.choice (['Male', 'M', 'female', 'Female', 'None'], size=1000),
       'Diagnosis': np.random.choice (['Heart attack','Diabetes', 'Flu','Stroke'], size=1000),
       'Treatment_Cost': np.random.uniform(500, 10000, size=1000),
       'Readmitted': np.random.choice(['Yes', 'No'], size=1000, p=[0.3, 0.7])
       }
df = pd.DataFrame(data)

#Export to CSV for our "Data Managment" records
df.to_csv('healthcare_raw_data.csv', index=False)
print("Success! 'healthcare_raw_data.csv' with 1000 rows is ready.")
```

```python
df.head()
```

```python
# 1. Round the cost to 2 decimal places
df['Treatment_Cost'] = df['Treatment_Cost'].round(2)

# 2. Fix the Gender mess (Standardizing)
# map the 'bad' names to 'good' names
gender_map = {'M': 'Male', 'female': 'Female', 'Male': 'Male', 'Female': 'Female'}
df['Gender'] = df['Gender'].map(gender_map).fillna('Others')

# 3. Fix the Age (Remove negative ages)
# "Only keep rows where age is 0 or older"
df = df[df['Age'] >= 0]

# 4. Drop the Name column (since it's not useful)
df = df.drop(columns=['Name'])

# Now let's see the clean version
df.head()
```
